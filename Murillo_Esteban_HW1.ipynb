```python
import pandas as pd
import numpy as np
```

```python
# Loading the dataset, reading the csv, and printing out the first 10 values of the DataFrame

myDF = pd.read_csv('wineQualityReds.csv', header = 0)
print(myDF.head(10))

```

```python
# sorting the values by acidity in descending order
print(myDF.sort_values(by = 'volatile.acidity', ascending = False))
```

```python
# finding and printing out wines with a quality of 7
seven_quality = myDF[myDF['quality'] == 7]

print(seven_quality)
```

```python
# Finding the average pH level across all wines
avg_ph = myDF["pH"].mean(skipna=True)
print("Average wine pH is: ", avg_ph)
```

```python
# filtering for wines whose alcohol level is greater than or equal to 10
ten_level = myDF[myDF['alcohol'] >= 10]

count = ten_level.shape[0]
print(count)
```

```python
# Finding the wine that contains the highest level of alcohol
max_wine = myDF['alcohol'].max()
print(max_wine)
```

```python
# Finding a random sample using dataframes
random_wine = myDF.sample(n=1, random_state=None)

# Extract its residual sugar level
residual_sugar = random_wine["residual.sugar"].values[0]

print(residual_sugar)
```

```python
# Filtering for wines with a quality of 4 and then finding a sample wine with that quality

quality_4 = myDF[myDF['quality'] == 4]
sample_wine_2 = myDF.sample(n = 1, random_state = None)

print(sample_wine_2)


```

```python
# Finding all wines that do not have a quality of 4, resetting the index, and then getting the count of the number of wines left
myDF = myDF[myDF["quality"] != 4]

print(myDF['quality'])

print(myDF)

# Reset the index (optional, but often useful after dropping rows)
myDF = myDF.reset_index(drop=True)

print("Remaining number of wines: ", myDF['Wine'].count())
```
